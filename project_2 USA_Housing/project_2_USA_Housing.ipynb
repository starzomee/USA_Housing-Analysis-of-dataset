# machine learning

```python
# split data to be used by the model 
# Create matrix of features 
x = full_data.drop('Price' , axis = 1)   # grabe all else but 'Price'

# Create target variable 
y = full_data['Price']  # y is the column to be predicted 
```

```python
x.shape
```

```python
y.shape
```

```python
# Use x and y variable to split the training and testing data into train and test set 
from sklearn.model_selection import train_test_split 
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = .20 , random_state = 101)
```

```python
x_train.shape
x_train
```

```python
y_train
```

```python
x_test.shape
x_test
```

```python
y_test
```

# linear Regression

# model Training

```python
# fit 
# import model 
from sklearn.linear_model import LinearRegression 

# Create instance of model 
lin_reg = LinearRegression()

# Pass training data into the model 
lin_reg.fit(x_train , y_train)
```

# model testing

# Class prediction

```python
# prediction 
y_pred = lin_reg.predict(x_test)
print(y_pred.shape)
print(y_pred)
```

```python
sns.scatterplot(x =y_test , y = y_pred , color = "blue" , label = "Actual Data Points ")
plt.plot([min(y_test) , max(y_test)] , [min(y_test) , max(y_test)] , color = "red"  , label = "Ideal line")
plt.legend()
plt.show()
```

```python
# combine actual and predicted values side by side 
results = np.column_stack((y_test , y_pred))
```

```python
# printing the results 
print("Actual values   |  Predicted values")
print('----' * 30)
for actual , predicted in results:
    print(f"{actual:14.2f}   | {predicted:12.2f}")
```

# Residual Anaylsis 

```python
residual = actual - y_pred.reshape(-1)
print(residual)
```

```python
# distribution plot for the residual analysis (difference between actaul and predicted )
sns.displot(residual , kde = True)
plt.show()
```

# Model Evaluation

```python
# Score it 
from sklearn.metrics import mean_squared_error 

print("Linear Regression Model ")
# results 
print("--"*30)
# mean_squared_error(y_test, y_pred)
mse = mean_squared_error(y_test , y_pred)
rmse = np.sqrt(mse)

# print evaluation matrics 
print("Mean Squared Error" , mse)
 
print("Root Mean Squared Error" , rmse)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

rf_regressor = DecisionTreeRegressor()
rf_regressor.fit(x_train , y_train)

# predicting the SalePrice using test set
y_pred_rf = rf_regressor.predict(x_test)

Dtr = mean_squared_error(y_pred_rf , y_test)

print("Decision Tree Regressor :" ,  Dtr)
```

# Random forest 

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

rf_regressor = RandomForestRegressor()
rf_regressor.fit(x_train , y_train)

# predicting the SalePrice using test set
y_pred_rf = rf_regressor.predict(x_test)

RFr = mean_squared_error(y_pred_rf , y_test)

print("Random Forest Regressor :" ,  RFr)
```

# Gradient Boosting Regression

```python
from sklearn.tree import DecisionTreeRegressor 
from sklearn.ensemble import RandomForestRegressor 
from sklearn.ensemble import GradientBoostingRegressor 

rf_regressor  = GradientBoostingRegressor()
rf_regressor.fit(x_train , y_train)

# predict 
y_pred_rf = rf_regressor.predict(x_test)
# random forest regressor accuracy check test 
GBr = mean_squared_error(y_pred_rf , y_test)
print("Gradient Boosting Regresstion :" , GBr)
```
